## 3. Sampleo
<p align="justify">
En este punto debes comprobar como se da la operación de sampleo y que sucede ante cambios en el intervalo de sampleo.    


### 3.1 Crea una señal impulso (delta) y desplazala
Para esto puedes usar alguna librería de python

```python
def dirac(n):
  return np.where(n == x, 1, 0)

x = 0 # Variable que permite elegir el momento del impulso
n = np.arange(-10, 11)
plt.stem(n, dirac(n))
plt.title('Impulso Unitario')
plt.xlabel('tiempo')
plt.ylabel('$\delta [t]$')
plt.ylim([-0.1, 1.1])
```

```python
def dirac(n):
  return np.where(n == x, 1, 0)

x = 3 # Al cambiar el valor de "x" se desplaza el impulso
n = np.arange(-10, 11)
plt.stem(n, dirac(n))
plt.title('Impulso Unitario')
plt.xlabel('tiempo')
plt.ylabel('$\delta [t]$')
plt.ylim([-0.1, 1.1])
```

### 3.2 Genere una señal Senoidal y realice la operación de sampleo
Puedes utilizar el método "stem" de matplotlib para plotear el resultado.

```python
def sensimple(t):
  return np.sin(t)

t = np.linspace(-5, 20, 100)
plt.plot(t, sensimple(t))
plt.title('f(t) = sen(t)')
plt.xlabel('tiempo')
plt.ylabel('Amplitud')
```

```python
muestreo = np.linspace(0, 20, 50) # se genera el intervalo de muestreo. En este intervalo se puede modificar el inicio, el final y la frecuencia a partir de la cantidad de elementos.
muestreo
```

```python
def dirac(n):
  return np.where(n == x, 1, 0)

def sensimple(t):
  return np.sin(t)

def sampleo(n, t):
  return dirac(n) * sensimple(t) # función que me devolverá las muestras dado que para impulso=1, sampleo=sen y para impulso=0, sampleo=0


x = muestreo
n = muestreo
t = muestreo
plt.stem(t, sampleo(n, t))
plt.title('g(t) = $\delta (n)$ * sin(t)')
plt.ylabel('Amplitud')
plt.xlabel('tiempo')
plt.xlim([-5, 20])
```

### 3.4 Realice la operación de sampleo sobre la señal que generaste en el punto 2

```python
def dirac(n):
  return np.where(n == x, 1, 0)

def señal(t):
  return np.sin(t)*2+np.cos(t*2)

def sampleo2(n, t):
  return dirac(n) * señal(t)


x = muestreo
n = muestreo
t = muestreo
plt.stem(t, sampleo2(n, t))
plt.title('g(t) = $\delta (n)$ * sin(t)')
plt.ylabel('Amplitud')
plt.xlabel('tiempo')
plt.xlim([-5, 20])
```
