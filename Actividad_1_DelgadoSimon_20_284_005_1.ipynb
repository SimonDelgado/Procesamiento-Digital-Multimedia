```python
import numpy as np
```

```python
array = np.linspace(0, 10, 50) # se crea un espacio de 50 elementos donde se evaluarán las funciones
array
```

```python
funcion1 = np.exp(-array**2) #función 1 con x=[0, 10]
funcion1
```

```python
def funcion2(z): #función 2, con x e y parametrizados en función de z
  x = z
  y = z
  return y*np.sin(x)

funcion2(array)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.title("Gráfica 1")
plt.plot(array, funcion1) # gráfica función 1
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
```

```python
plt.title("Gráfica 2")
plt.plot(array, funcion2(array)) #gráfica función 2
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
```

### 3.1 Ruido Gaussiano
Debes agregar la señal de ruido a la que creaste previamente. Adicionalmente, debes mostrar una gráfica con el resultado.


```python
aleatorio = np.random.random(50) #se crea vector aleatorio para fórmula gaussiana
aleatorio
```

```python
def gauss(x, mu, sigma): #se genera señal de ruido
  return (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * (((x - mu) ** 2) / sigma))

gauss(aleatorio, 0, 1) # se presenta una muestra con distribución uniforme de máximo 1
```

```python
funcion = np.sin(2*array)+np.cos(array) # funcion pura sin ruido agregado
funcion
```

```python
plt.title("Gráfica 3")
plt.plot(array, funcion) #gráfica de función sin ruido
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
```

```python
funcionruido1 = funcion + gauss(aleatorio, 0, 1)
funcionruido1
```

```python
plt.title("Gráfica 4")
plt.plot(array, funcionruido1) # la primera gráfica otorga una deformación leve de la señal
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
```

```python
funcionruido2 = funcion + gauss(aleatorio, 1, 0.5) # se utilizó la combinación de mu=1 y sigma=0.5 ya que la gráfica otorga una mejor visualización del ruido
```

```python
plt.title("Gráfica 5")
plt.plot(array, funcionruido2)
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
```
